```python
import numpy as np
import matplotlib.pyplot as plt

import lingam
```

```python
DATA_NUM = 1000
def generate_data():
  np.random.seed(777)
  x_2 = np.random.rand(DATA_NUM) * 4 - 2
  e_1 = np.random.rand(DATA_NUM) * 4 - 2
  x_1 = 1.1 * x_2 + e_1
  return x_1, x_2

def scatter(x: float, x_label: str, y: float, y_label: str):
  fig = plt.figure()
  ax = fig.add_subplot(111)
  ax.set_xlabel(x_label)
  ax.set_ylabel(y_label)
  ax.scatter(x, y)
  plt.show()
```

```python
x_1, x_2 = generate_data()
```

```python
scatter(x_1, "x_1", x_2, "x_2")
```

```python
scatter(x_2, "x_2", x_1, "x_1")
```

```python
train_data = np.r_["1", x_1.reshape(DATA_NUM,1), x_2.reshape(DATA_NUM,1)]
model = lingam.DirectLiNGAM()
model.fit(train_data)
```
